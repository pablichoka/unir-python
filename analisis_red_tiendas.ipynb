```python
import pandas as pd
import numpy as np
print("Librerías importadas correctamente.")
```

```python
df_ventas = pd.read_csv('ventas.csv')
df_inventarios = pd.read_csv('inventarios.csv')
df_satisfaccion = pd.read_csv('satisfaccion.csv')

df_ventas.dropna(inplace=True)
df_inventarios.dropna(inplace=True)
df_satisfaccion.dropna(inplace=True)

print("Datos de Ventas:")
display(df_ventas.head())

print("\nDatos de Inventarios:")
display(df_inventarios.head())

print("\nDatos de Satisfacción:")
display(df_satisfaccion.head())
```

```python
productos = df_ventas.groupby('Producto')['Cantidad_Vendida'].sum()
print(productos)

tiendas = df_ventas.groupby('ID_Tienda')['Cantidad_Vendida'].sum()
print(tiendas)

df_ventas['Ingresos'] = df_ventas['Precio_Unitario'] * df_ventas['Cantidad_Vendida']
print(df_ventas)

ingreso_tienda = df_ventas.groupby('ID_Tienda')['Ingresos'].sum()
print(ingreso_tienda)
```

```python
estadisticas_venta = df_ventas.describe()
print(estadisticas_venta)

cantidad_media_vendida = df_ventas.groupby(['ID_Tienda', 'Producto'])['Cantidad_Vendida'].mean()
print(cantidad_media_vendida)
```

```python
ventas_agregadas = df_ventas.groupby(['ID_Tienda', 'Producto'])['Cantidad_Vendida'].sum().reset_index()
print(ventas_agregadas)

df_combinado = pd.merge(df_inventarios, ventas_agregadas, on=['ID_Tienda', 'Producto'], how='inner')
print("\nDataFrame Combinado (Inventarios + Ventas Agregadas):")
print(df_combinado.head())
```

```python
df_combinado['Rotacion_Inventario'] = df_combinado['Cantidad_Vendida'] / df_combinado['Stock_Disponible']
print(df_combinado)

tiendas_criticas = df_combinado[df_combinado['Rotacion_Inventario'] < 0.1]
print("\nTiendas y Productos con Rotación de Inventario Crítica (< 10%):")
print(tiendas_criticas)
```

```python
df_definitivo = pd.merge(df_combinado, df_satisfaccion, on=['ID_Tienda'], how='inner')
print(df_definitivo)

tiendas_mejorables = df_definitivo[df_definitivo['Satisfacción_Promedio'] < 60]
print(tiendas_mejorables)
```

```python
ventas_totales = df_ventas['Ingresos'].to_numpy()
print(ventas_totales)

print(np.median(ventas_totales))

print(np.std(ventas_totales))

np.random.seed(42)

proyeccion_ventas = np.random.randint(1000, 15000, 10)
print(proyeccion_ventas)
```
